```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('netflix_titles.csv')
```

```python
df.head(30)
```

```python
info_df = df.info()

print(info_df)
```

```python
df.columns
```

```python
eua = df[df['country'] == 'United States']

print(eua)
```

```python
brazil = df[df['country'] == 'Brazil']

print(brazil)
```

```python
japan = df[df['country'] == 'Japan']

print(japan)
```

```python

paises_filtrados = (eua, japan, brazil)
df_filtrado = df[df[paises_filtrados]]

contagem_paises = df_filtrado.value_counts()

contagem_paises.plot.bar(kind='bar', figsize=(10, 6), color=['skyblue', 'salmon', 'lightgreen'])


# Os rotulos
plt.xlabel('Países')
plt.xticks(rotation=0) # colocando o rotulo do eixo X deitado
plt.ylabel('Frequência')
plt.title('Quantidade de Programas por País')
#plt.legend(['Taxa de Sobrevivencia'])


```
